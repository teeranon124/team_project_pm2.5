```python
from pycaret.datasets import get_data
data = get_data('insurance')
```

```python
from pycaret.regression import RegressionExperiment
exp1 = RegressionExperiment()
exp1.setup(data, target='charges', session_id=123)
 
```

```python
best = exp1.compare_models()
```

```python
exp1.evaluate_model(best)
```

```python
pred_holdout = exp1.predict_model(best)
```

```python
new_data = data.copy().drop(['charges'], axis=1)
predictions = exp1.predict_model(best, data=new_data)
predictions.head()
```

```python
exp1.save_model(best, 'models/best_model')
```
